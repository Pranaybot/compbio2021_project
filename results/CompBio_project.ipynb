```python
import pandas as pd
import os
import seaborn as sb
```

```python
bromo_dt={}
directory=r'/Users/pranaypentaparthy/Desktop/bromodomain_protein_folder/'
for filename in os.scandir(directory):
    protein_data = pd.read_csv(filename, sep='\t')
    protein_data = protein_data[['Accession', 'Name']].assign(Protein=filename)
    bromo_dt[filename] = protein_data
```

```python
bromo_df = pd.DataFrame()
for key in bromo_dt.keys():
    data = bromo_dt[key]
    bromo_df = bromo_df.append(data)
bromo_df.to_csv('/Users/pranaypentaparthy/Desktop/CMSE 410/protein_features.tsv', index=False, sep='\t')
```
